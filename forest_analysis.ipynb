```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
df=pd.DataFrame()
df=pd.read_csv('forest.csv', skiprows=4)
df1=df.ix[:,'1990':'2011']
df2 = df1.dropna(thresh=19)
fill_value = pd.DataFrame({col: df2.mean(axis=1) for col in df2.columns})
df2.fillna(fill_value, inplace=True)
df2['avg_forestcover'] = df2.mean(axis=1)
df3 = df2.sort('avg_forestcover', ascending=True)
df3['country_name'] = df.ix[:,'Country Name']
d1 = df3[df3['country_name'].str.contains('income')]
d2 = df3[df3['country_name'].str.contains('countries')]

d11=d1.ix[:,'1990':'1995']
d11['avg_forestcover'] = d11.mean(axis=1)
d11['country_name'] = d1.ix[:,'country_name']

d12=d1.ix[:,'1996':'2000']
d12['avg_forestcover'] = d12.mean(axis=1)
d12['country_name'] = d1.ix[:,'country_name']

d13=d1.ix[:,'2001':'2005']
d13['avg_forestcover'] = d13.mean(axis=1)
d13['country_name'] = d1.ix[:,'country_name']

d14=d1.ix[:,'2006':'2011']
d14['avg_forestcover'] = d14.mean(axis=1)
d14['country_name'] = d1.ix[:,'country_name']

d21=d2.ix[:,'1990':'1995']
d21['avg_forestcover'] = d21.mean(axis=1)
d21['country_name'] = d2.ix[:,'country_name']

d22=d2.ix[:,'1996':'2000']
d22['avg_forestcover'] = d22.mean(axis=1)
d22['country_name'] = d2.ix[:,'country_name']

d23=d2.ix[:,'2001':'2005']
d23['avg_forestcover'] = d23.mean(axis=1)
d23['country_name'] = d2.ix[:,'country_name']

d24=d2.ix[:,'2006':'2011']
d24['avg_forestcover'] = d24.mean(axis=1)
d24['country_name'] = d2.ix[:,'country_name']

d11=d11.set_index(['country_name'])
d11.plot(x=d11.index,kind='bar')
#ax.set_xticklabels(d11.index, rotation=180)
plt.xlabel('Country categories ')
plt.ylabel('Forest Cover % ')
plt.show()

d14=d14.set_index(['country_name'])
d14.plot(x=d14.index,kind='bar')
#ax.set_xticklabels(d11.index, rotation=180)
plt.xlabel('Country categories ')
plt.ylabel('Forest Cover % ')
plt.show()


d21=d21.set_index(['country_name'])
d21.plot(x=d21.index,kind='bar')
#ax.set_xticklabels(d11.index, rotation=180)
plt.xlabel('Country categories ')
plt.ylabel('Forest Cover % ')
plt.show()

d24=d24.set_index(['country_name'])
d24.plot(x=d24.index,kind='bar')
#ax.set_xticklabels(d11.index, rotation=180)
plt.xlabel('Country categories ')
plt.ylabel('Forest Cover % ')
plt.show()


df3=df3.set_index(['country_name'])
dff=df3.ix[:,'20011':]
dff1=dff.tail(10)
dff1.head(10)
```
